```python
import pickle
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
model_name = 'e2f_final/x_True_True3/weights/epoch16'
window_size = 5

try:
    with open(model_name+'_gt_objects_val.pkl', 'rb') as f:
        gt_obj = pickle.load(f)
    with open(model_name+'_pred_objects_val.pkl', 'rb') as f:
        pred_obj = pickle.load(f)
    avg_gt = []
    avg_pred = []
    total_score = []
    for i in range(len(gt_obj)):
        gt_obj_avg = np.convolve(gt_obj[i], np.ones(window_size)/window_size, mode='valid')
        pred_obj_avg = np.convolve(pred_obj[i], np.ones(window_size)/window_size, mode='valid')
        avg_gt.append(gt_obj_avg)
        avg_pred.append(pred_obj_avg)
        score = np.sum(np.array(pred_obj_avg) - np.array(gt_obj_avg))/len(pred_obj_avg)
        total_score.append(round(score, 2))
        plt.figure()
        plt.plot(avg_gt[-1], label = 'conventional camera predictions')
        plt.plot(avg_pred[-1], label = 'event camera predictions')
        plt.legend()
        plt.title(f'Moving Average Object Count on test data {i}')
        plt.xlabel('Frames')
        plt.ylabel('Number of objects')
        plt.show()
        plt.close()
    print(model_name)
    print(total_score, end = '  ') 
    # if np.sum(total_score) > 25:
    print(round(np.sum(total_score), 2))
    print()
except FileNotFoundError as e:
    print(e)
    pass
```

```python
with open(model_name+'gt_objects_val_extended.pkl', 'rb') as f:
    gt_obj = pickle.load(f)
with open(model_name+'pred_objects_val_extended.pkl', 'rb') as f:
    pred_obj = pickle.load(f)
avg_gt = []
avg_pred = []
total_score = []

before_gts = []
before_preds = []
at_gts = []
at_preds = []
after_gts = []
after_preds = []
for i in range(len(gt_obj)):
    gt_obj_avg = np.convolve(gt_obj[i], np.ones(window_size)/window_size, mode='valid')
    pred_obj_avg = np.convolve(pred_obj[i], np.ones(window_size)/window_size, mode='valid')
    before_gt = sum(gt_obj_avg[:45])/45
    before_pred = sum(pred_obj_avg[:45])/45
    at_gt = sum(gt_obj_avg[45:-60])/(len(gt_obj_avg)-105)
    at_pred = sum(pred_obj_avg[45:-60])/(len(gt_obj_avg)-105)
    after_gt = sum(gt_obj_avg[-60:])/60
    after_pred = sum(pred_obj_avg[-60:])/60
    before_gts.append(before_gt)
    before_preds.append(before_pred)
    at_gts.append(at_gt)
    at_preds.append(at_pred)
    after_gts.append(after_gt)
    after_preds.append(after_pred)
    # plt.plot([before_gt, at_gt, after_gt])
    # plt.plot([before_pred, at_pred, after_pred])

means = [np.mean(np.array(before_gts)-np.array(before_preds)), 
         np.mean(np.array(at_gts)-np.array(at_preds)), 
         np.mean(np.array(after_gts)-np.array(after_preds))]
stds = [np.std(np.array(before_gts)-np.array(before_preds)), 
         np.std(np.array(at_gts)-np.array(at_preds)), 
         np.std(np.array(after_gts)-np.array(after_preds))]
plt.errorbar([0, 1, 2], means, yerr=stds, fmt='o', capsize=5, label="Mean ± Std")
plt.scatter([0,1,2],means)
plt.title(f'Objects before at and after {i}')
plt.show()
plt.close()



```
